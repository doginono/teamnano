```python
import numpy as np

spot = spot.sort_values(  
    "time"
).reset_index(  
    drop=True
)

spot["mid_price"] = (spot["bid_price"] + spot["ask_price"]) / 2  
```

```python
t_ns = spot["time"].astype("datetime64[ns]").values.astype(np.int64)
dt_ns = 3_000_000  # 3 ms in nanoseconds

# --- end-index of the 3 ms window for every row
window_end = np.searchsorted(t_ns, t_ns + dt_ns, side="right")

mid_price = spot["mid_price"].to_numpy()

# --- pre-allocate arrays for future max/min
max_future = np.empty_like(mid_price)
min_future = np.empty_like(mid_price)

for i in range(len(mid_price)):
    j = window_end[i]  # first row **after** the 3 ms horizon
    if j > i + 1:  # at least one future row exists
        slice_ = mid_price[i + 1 : j]
        max_future[i] = slice_.max()
        min_future[i] = slice_.min()
    else:  # no data inside the 3 ms window
        max_future[i] = mid_price[i]
        min_future[i] = mid_price[i]
```

```python
rel_up = (max_future - mid_price) / mid_price  # positive move
rel_down = (mid_price - min_future) / mid_price  # negative move

spot["shock"] = np.where(
    rel_up >= 0.0007,
    1,  # upward shock
    np.where(rel_down >= 0.0007, -1, 0),  # downward shock
)  # no shock
```

```python
fut = fut.sort_values("time").reset_index(drop=True)

# add mid-prices for futures too
fut["mid_price"] = (fut["bid_price"] + fut["ask_price"]) / 2
```

```python
t_spot_ns = spot["time"].astype("datetime64[ns]").values.astype(np.int64)
t_fut_ns = fut["time"].astype("datetime64[ns]").values.astype(np.int64)
```

```python
# 5 ms in nanoseconds
horizon_ns = 5_000_000

# idx_first_future[i] = position in `futures` DF where time >= t_spot[i]
idx_first_future = np.searchsorted(t_fut_ns, t_spot_ns, side="left")

# idx_horizon[i]     = first row **after** t_spot[i] + 5 ms
idx_horizon = np.searchsorted(t_fut_ns, t_spot_ns + horizon_ns, side="left")
```

```python
# empty column – we’ll fill only for shocks
spot["fut_follow"] = np.nan

thr = 0.0002  # 0.02 % threshold to ignore micro noise – change to taste

for i, row in spot[spot["shock"] != 0].iterrows():

    j0 = idx_first_future[i]  # 1st futures tick at or after shock time
    j1 = idx_horizon[i]  # 1st tick AFTER the 5 ms window

    if j0 == j1:  # no futures data in the window → NaN
        continue

    # baseline = futures mid **just before** the shock (or same tick)
    if j0 == 0:
        baseline = fut["mid_price"].iloc[j0]
    else:
        baseline = fut["mid_price"].iloc[j0 - 1]

    # reaction = first futures mid inside the window
    reaction = fut["mid_price"].iloc[j0]

    rel_change = (reaction - baseline) / baseline

    if rel_change >= thr:
        label = +1
    elif rel_change <= -thr:
        label = -1
    else:
        label = 0

    spot.at[i, "fut_follow"] = label
```

```python
# hit-rate: fraction of shocks that futures followed in same direction
hits = spot[
    (spot["shock"] == spot["fut_follow"])
    & (spot["shock"] != 0)
    & (~spot["fut_follow"].isna())
]
hit_rate = len(hits) / len(spot[spot["shock"] != 0])
print(f"Lead-success rate (spot → futures): {hit_rate:.2%}")

# distribution of follow labels
print(spot["fut_follow"].value_counts(dropna=False))
```

```python
trades = trades.sort_values("time").reset_index(drop=True)
```

```python
window_ms = 10
window_ns = window_ms * 1_000_000
```

```python
t_spot_ns = spot["time"].astype("datetime64[ns]").values.astype(np.int64)
t_trade_ns = trades["time"].astype("datetime64[ns]").values.astype(np.int64)
q_trade = trades["quantity"].to_numpy()
```

```python
idx_left = np.searchsorted(t_trade_ns, t_spot_ns - window_ns, side="left")
idx_right = np.searchsorted(t_trade_ns, t_spot_ns + window_ns, side="right")
```

```python
spot["trade_vol"] = 0.0  # initialise

shocks_idx = spot.index[spot["shock"] != 0]  # rows with ±1 shock

for i in shocks_idx:
    # total_qty = q_trade[idx_left[i] : idx_right[i]].sum()
    total_qty = q_trade[
        idx_left[i] : np.searchsorted(t_trade_ns, t_spot_ns[i], side="left")
    ].sum()
    spot.at[i, "trade_vol"] = total_qty
```

```python
threshold = spot.loc[shocks_idx, "trade_vol"].median()  # median split

spot["vol_bucket"] = np.where(spot["trade_vol"] >= threshold, "high", "low")
```

```python
def lead_success(df):
    hits = df[(df["shock"] == df["fut_follow"]) & (~df["fut_follow"].isna())]
    return len(hits) / len(df) if len(df) else np.nan

high_idx = shocks_idx.intersection(spot.index[spot["vol_bucket"] == "high"])
low_idx = shocks_idx.intersection(spot.index[spot["vol_bucket"] == "low"])

high_df = spot.loc[high_idx]
low_df = spot.loc[low_idx]
print(f"Threshold qty : {threshold:.2f}")
print(f"High-vol shocks: {len(high_df):>6}  | hit-rate: {lead_success(high_df):.2%}")
print(f"Low-vol  shocks: {len(low_df):>6}  | hit-rate: {lead_success(low_df):.2%}")
```

```python
print(
    spot.loc[shocks_idx, "trade_vol"].describe(
        percentiles=[0.25, 0.5, 0.75, 0.9, 0.95, 0.99]
    )
)
```

```python
high_vol=[]
low_vol=[]
for i in [0.02* j for j in range(1, 50)]:
    threshold = spot.loc[shocks_idx, "trade_vol"].quantile(i)  # top 10%
    print(f"New threshold: {threshold:.2f}")
    high_idx = shocks_idx.intersection(spot.index[spot["trade_vol"] >= threshold])
    low_idx = shocks_idx.intersection(spot.index[spot["trade_vol"] < threshold])

    high_df = spot.loc[high_idx]
    low_df = spot.loc[low_idx]

    print(f"Threshold qty : {threshold:.2f}, percentile: {i:.2%}")
    print(f"High-vol shocks: {len(high_df):>6}  | hit-rate: {lead_success(high_df):.2%}")
    print(f"Low-vol  shocks: {len(low_df):>6}  | hit-rate: {lead_success(low_df):.2%}")
    high_vol.append(lead_success(high_df))
    low_vol.append(lead_success(low_df))
```

```python
best_idx = int(np.argmax(high_vol))  # position in the list
best_percentile = (best_idx + 1) * 0.02  # 0.02 → 0.98 (because index 0 == 2 %)

print(f"Best percentile : {best_percentile:.0%}")  # e.g. 68 %
# -------------------------------------------------
# 2) Get the corresponding volume threshold
#     (use only shock rows for the quantile)
# -------------------------------------------------
trade_series = spot.loc[shocks_idx, "trade_vol"]
best_threshold = trade_series.quantile(best_percentile)

print(f"Best volume threshold : {best_threshold:.2f} TRB")

# -------------------------------------------------
# 3) Tag every row with 'high' or 'low' volume bucket
# -------------------------------------------------
spot["vol_bucket"] = np.where(
    spot["trade_vol"] >= best_threshold,
    "high",
    "low",
)
```
